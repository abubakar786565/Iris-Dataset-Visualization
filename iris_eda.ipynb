# Task 1: Exploring and Visualizing a Simple Dataset

- *Import required libraries*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- *Load the Iris dataset from seaborn*

```python
iris = sns.load_dataset("iris")
```

- *Basic Inspection*

```python
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows:")
print(iris.head())
```

- *Summary Information*

```python
print("\nDataset Info:")
print(iris.info())

print("\nStatistical Summary:")
print(iris.describe())
```

- *Scatter Plot: Feature Relationships*

```python
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

- *Histograms for Each Feature*

```python
iris.hist(figsize=(10, 8), edgecolor="black")
plt.suptitle("Histogram of Each Feature", y=1.02)
plt.tight_layout()
plt.show()
```

- *Box Plots to Identify Outliers*

```python
plt.figure(figsize=(12, 6))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=iris[column])
    plt.title(f"Boxplot of {column}")

plt.tight_layout()
plt.suptitle("Box Plots to Identify Outliers", y=1.02)
plt.show()
```
